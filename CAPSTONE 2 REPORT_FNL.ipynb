Let's review the baseline for the experiment - what options owners provide as a delivery

```python
df ['fee']
```

```python
# Cleaning the data for the FEE , 
df['fee'].replace(regex=True,inplace=True,to_replace=r'\$', value=r'')
df['fee'].replace(regex=True,inplace=True,to_replace=r'Up to', value=r'')
df['fee'].replace(regex=True,inplace=True,to_replace=r'£', value=r'')
df['fee'].replace(regex=True,inplace=True,to_replace=r'NaN', value=r'-1')
```

```python
#Cleaning Nan to 0 - no delivery
df['fee'].fillna(0, inplace=True)
```

After all the cleaning let's categorize the data so it'll be easier to build the baseline :

```python
#replace 'free' into int
df['fee'] = df['fee'].replace(to_replace =['Free'],
                  value = '-999')
```

```python
#convert to Int
df['fee'] = df['fee'].astype(int)
```

```python
# Create a list to store the data
fee_cat = []

# For each row in the column,
for row in df['fee']:
    # if more than a value,
    if row > 0:
        # Append a letter grade
        fee_cat.append('Paid Delivery')
    # else, if free,
    elif row < 0:
        # Append a letter grade
        fee_cat.append('Free')
    # otherwise,
    else:
        # Append a failing grade
       fee_cat.append('No delivery')
        
# Create a column from the list
df['fee_cat'] = fee_cat
```

```python
df['fee_cat']
```

Let's move to the next stage - calculate the Utilization rate - the key indicator of our successful or not experiment  - #Utilisation rate - % of month that car has been rented

```python
df['UR'] = ((df['Occupancy Mar 18']+2 - df['Unavailable by owner Mar18']))/((31-df['Unavailable by owner Mar18']))
#remove null
df = df[pd.to_numeric(df['UR'], errors='coerce').notnull()]
df['UR']
```

```python
#Drop all values that aren't clean 
#remove all negative
df = df[(df[['UR']] > 0).all(1)]
#remove a inll more than 1
df = df[(df[['UR']] < 1).all(1)]
#drop all inf
df.replace([np.inf, -np.inf], np.nan).dropna(subset=["UR"], how="all")
```

```python
#TOP 10 cars by earnings/month in Mar 2018 in every category
df['Earning'] = df['UR'] * 31*df['Price per day']
dfs = df [['Make', 'Model','Price per day', 'UR','Category', 'Earning', 'fee_cat']]
dfs.groupby(['Category']).apply(lambda x: x.nlargest(5, columns=['Earning']))
```

Based on the chart above we can say that the TOP 5 cars within a category has paid delivery. Let's confirm it for the whole set of the data with the pointplot

```python
sns.set(style="whitegrid")

g = sns.pointplot(x="Category", y="UR", hue= 'fee_cat', data=df,
                   height=6, kind="strip", palette="pastel",ci=95,dodge=True,join=False)
plt.title('UR by Category')
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right', fontsize='x-small')
plt.legend(loc='upper right')
plt.show()
```

This will be enough data for us to conclude that the delivery has a significant value for the UR and earnings.
